### Model function

Let's try to get a better intuition for this through the code blocks below. 

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear module
    It gives the Y-array of the passed x
    """
    m = x.shape[0]
    # creating the returning array and filling it with zeros
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

Ploting the output using the `x_train` values and the `compute_model_output` function, to test the `w` and the `b` values of choise

```python
tmp_f_wb = compute_model_output(x_train,w,b)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, c='r', marker='x')
plt.title('Housing Prices')
plt.ylabel('Price (1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.plot()
```
